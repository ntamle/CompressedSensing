## 2-D Wavelet transform

### Haar basis in $L^2([0,1])$

Wavelet basis are constructed from a **wavelet mother** like the Haar function: for all $0\leq t\leq1$,

$$\Psi(t)=\left\{
\begin{array}{cc}
1 & \mbox{ if }  0\leq t<1/2\\
-1 & \mbox{ if } 1/2 \leq t <1\\
0 & otherwise
\end{array}
\right.$$ and a **scaling function** $\phi(t)=1$.


Wavelet basis functions are of the form

$$\Psi_i^j(x)=\sqrt{2^j}\Psi(2^j x - i ), \quad i=0,1,\cdots, 2^j-1; j=0,1,\ldots,$$ 

and the Haar basis of $L^2([0,1])$ is $\{\phi\}\cup\{\Psi_i^j: i=0,1,\cdots, 2^j-1; j=0,1,\ldots,\}$.


### Haar basis in $L^2(\{0, \cdots, n-1\}^2)$

We use only the first low-frequency wavelet functions to construct a wavelet basis of $L^2(\{0, \ldots, n-1\}^2)$:

$$\Psi_i^j(x)=\sqrt{2^j}\Psi(2^j x - i ), \quad i=0,1,\cdots, 2^j-1; j=0,1,\cdots, \log_2 n-1$$


We denote $w(0, x)=\phi(x)$, $w(1,x)=\Psi(x)$ and for all $j=1,\cdots, \log_2 n-1$ and $i=0,1,\cdots, 2^j-1$,

$$w(2j+i, x) = \Psi_i^j(x)$$

$(w(k,\cdot):k=0,\ldots,n)$ is the Haar basis of $L^2(\{0, \ldots,n-1\})$. 

The Haar basis of $L^2(\{0, \ldots,n-1\}^2)$ may be obtained from the Haar basis in $L^2(\{0, \ldots,n-1\})$ via tensorization: $(w(k,\cdot)\otimes w(m,\cdot):k, m=0,\ldots,n)$ is the Haar basis in $L^2(\{0, \ldots,n-1\})$ where

$$(w(k,\cdot)\otimes w(m,\cdot))(i,j) = w(k,i)w(m,j)$$

The wavelet transform of an image $f\in\{0,1,\cdots, n-1\}^2$ is for any $(x,y)\in\{0,\cdots,n-1\}^2$

$$ f(x,y) = \sum_{k=0}^{n-1}\sum_{m=0}^{n-1} fw(k,m) \times w(k,x) \times w(m,y)$$
where for any $(k,m)\in\{0,\cdots,n-1\}^2$

$$ fw(k,m) = \sum_{x=0}^{n-1}\sum_{y=0}^{n-1} f(x,y) \times w(k,x) \times w(m,y)$$

### Load a gray-scale image $f$ of $n \times n$ pixels

```python
n = 256
name = 'nt_toolbox/data/lena.png'
f = load_image(name, n)
```

```python
imageplot(f, 'Original image')
```

### Implement a full wavelet transform

```python
h = [0, .482962913145, .836516303738, .224143868042, -.129409522551] 
h = h/norm(h)
u = power(-ones(len(h)-1),range(1,len(h))) # alternate +1/-1
g = concatenate(([0], h[-1:0:-1] * u))
Jmin = 0
Jmax = log2(n)-1
fW = f.copy()
for j in arange(Jmax,Jmin-1,-1):
    A = fW[:int(2**(j+1)):,:int(2**(j+1)):]
    for d in arange(1,3):
        Coarse = subsampling(cconv(A,h,d),d)
        Detail = subsampling(cconv(A,g,d),d)
        A = concatenate( (Coarse, Detail), axis=d-1 )
    fW[:int(2**(j+1)):,:int(2**(j+1)):] = A
    j1 = Jmax-j
```

Check for orthogonality of the transform (conservation of energy).

```python
print( 'Energy of the signal/coefficients = ' + str(norm(f)/norm(fW)) )
#figure(figsize = (10,8))
#plot_wavelet(fW,Jmin);
```

## Inverse 2-D Wavelet transform

Inversing the wavelet transform means retrieving a signal $f1 = \big(f1(x,y):0\leq x,y\leq n-1\big)$ from the
coefficients $fw = \big(fw(k,m): 0\leq k,m\leq n-1\big)$. If $fw$ are exactely the coefficients of $f$, then
$f=f1$ up to machine precision. 

```python
f1 = fW.copy()
for j in arange(Jmin,Jmax+1): 
    A = f1[:int(2**(j+1)):,:int(2**(j+1)):]
    for d in arange(1,3):
        if d==1:
            Coarse = A[:int(2**j):,:]
            Detail = A[int(2**j):int(2**(j+1)):,:]
        else:
            Coarse = A[:,:int(2**j):]
            Detail = A[:,int(2**j):int(2**(j+1)):]
        Coarse = cconv(upsampling(Coarse,d),reverse(h),d)
        Detail = cconv(upsampling(Detail,d),reverse(g),d)
        A = Coarse + Detail
        j1 = Jmax-j
    f1[:int(2**(j+1)):,:int(2**(j+1)):] = A
```

Check that we recover exactly the original image.

```python
print( 'Error |f-f1|/|f| = ' + str(norm(f-f1)/norm(f)) )
```

## Linear 2-D Wavelet Approximation

Linear approximation is performed by **setting to zero the 'fine scale' or 'high resolution' wavelets coefficients**
and then performing the inverse wavelet transform.

Here we keep only $1/16$ of the wavelet coefficients -- that is $256^2/16 = 4.096$ coefficients compare with the total number of $65.536$ pixels.

```python
fW = perform_wavortho_transf(f,Jmin,+1,h)
eta = 4
fWLin = zeros((n,n))
fWLin[:int(n/eta):,:int(n/eta):] = fW[:int(n/eta):,:int(n/eta):]
fLin = perform_wavortho_transf(fWLin,Jmin,-1,h)
```

```python
plt.figure(figsize = (15,12))
plt.subplot(121)
imageplot(f, 'Original image 65.536 coefficients')
plt.subplot(122)
imageplot(clamp(fLin), 'Compressed image 4.096 coefficients')
plt.show()
```

## Threshold 2-D Wavelet Approximation

A non-linear $m$-term approximation is obtained by **keeping only the $m$
largest coefficients**, which creates the smallest possible error.


Removing the smallest coefficients and so keeping only the $m$-largest ones, is
equivalently obtained by thresholding small coefficient to $0$. 


First select a threshold value (the largest the threshold, the more
agressive the approximation).

```python
T = .2
```

Then set to 0 coefficients with magnitude below the threshold.

```python
fWT = fW * (abs(fW)>T)
```

Display thresholded coefficients.

```python
#representation of wavelest coefficient of patches with different sizes
#figure(figsize = (15,12))
#subplot(121)
#plot_wavelet(fW,Jmin);
#subplot(122)
#plot_wavelet(fWT,Jmin);
```

Plot the (log of) the coefficient in decaying order.

```python
# 256**2 = 65536
v = reverse( msort(abs(fW).ravel()))
figure(figsize = (15,6))
subplot(121)
plot(log10(v))
axis('tight');
plt.title('log of wavelets coefficients in decaying order')
subplot(122)
thres = 5000
vt = v[0:thres]
plot(log10(vt))
axis('tight');
plt.title('log of the first {} wavelets coefficients in decaying order'.format(thres))
#plt.savefig("log_wavelet_coef.png", bbox_inches='tight')
```

Find the thresholds $T$ so that the number $m$ of remaining coefficients in the threshold coefficients
are $m=n^2/16$. 

```python
eta = 4 #we keep (100/eta^2) percent of the coefficients
m = round((n**2)/(eta**2))
T = v[m]
```

Display the corresponding non-linear approximation.

```python
fWT = fW * (abs(fW)>=T)
fnl = perform_wavortho_transf(fWT,Jmin,-1,h)
```

```python
plt.figure(figsize = (15,12))
plt.subplot(121)
imageplot(f, 'Original image {} coefficients'.format(n**2))
plt.subplot(122)
imageplot(clamp(fnl), 'Compressed image {} coefficients'.format(m))
plt.savefig('comparaison_images_lena.png', bbox_inches='tight')
```
