```python
import scipy
from scipy import misc
import numpy as np
from PIL import Image
import time
import matplotlib.pyplot as plt


from numpy.linalg import svd


from cvxpy import *
print(installed_solvers())
%matplotlib inline
```

```python
# IMAGE LECTURE and RESCALING 

ascent = misc.ascent()
face = misc.face()

plt.figure(figsize = (10,10))
plt.subplot(1, 2, 1)
plt.imshow(ascent, cmap=plt.cm.gray)
plt.subplot(1, 2, 2)
plt.imshow(face)


face = face[:,:,0]
print(ascent.shape)

```

```python
#Transfo image svd 

face = face[0:768, 0:768] # to get a square matrix
U, s, V = svd(face)

#Truncated SVD and noise adding (testing purpose)

def truncated_svd(image = face, rank = 50):
    U, s, V = svd(image)
    n = s.size
    s[rank-n:-1] = 0
    "TODO: return the truncated svd of a matrix"
    return U, s, V  

def mask(u, v, proportion = 0.2):
    mat_mask = np.random.binomial(1, proportion, size =  (u, v))
    print("We observe {} per cent of the entries of a {}*{} matrix".format(100 * mat_mask.mean(),u, v))
    return mat_mask
```

```python
#OPTMIZATION ALGORITHM (1st ATTEMPT)

def nuclear_norm_mini(image):
    u, v = image.shape
    A = Variable((u, v))
    S = Variable((u, v))
    obj = Minimize(norm(A, 'nuc')+(1/u**(1/2))*norm(S, 1))
    constraints = [A + S == image]
    prob = Problem(obj, constraints)
    prob.solve(solver=SCS)
    if prob.status != 'optimal':
        print('CVXPY failed to reach optimal value')
    return A.value,S.value
```

```python
#FIRST RESULTS (NOT SATISFYING...)

rank = 50
size = 500
image = ascent[0:size, 0:size]
U, sr, V = truncated_svd(image, rank)
S_r = np.diag(sr)
image_r = np.dot(U, np.dot(S_r, V))
proportion = 0.9
u, v = image_r.shape
mat_mask = mask(u, v, proportion)

now_0 = time.time()
A, S= nuclear_norm_mini(multiply(mat_mask,image_r))
print("{} seconds".format(time.time() - now_0))


#PLOTTING 

plt.figure(figsize = (15, 10))
plt.subplot(1,3,1)
plt.imshow(image_r, cmap=plt.cm.gray)
plt.title("original image (rank = {})".format(rank))
plt.subplot(1,3,2)
plt.imshow(A, cmap=plt.cm.gray)
plt.title("LOW RANKING IMAGE (proportion = {} per cent)".format(proportion))
plt.subplot(1,3,3)
plt.imshow(S, cmap=plt.cm.gray)
plt.title("SPARSE IMAGE")
```

```python
# ITERATIVE ALGORITHM (GIVEN BY the article)

class R_pca:

    def __init__(self, D, mu=None, lmbda=None):
        self.D = D
        self.S = np.zeros(self.D.shape)
        self.Y = np.zeros(self.D.shape)

        if mu:
            self.mu = mu
        else:
            self.mu = np.prod(self.D.shape) / (4 * self.frobenius_norm(self.D))

        self.mu_inv = 1 / self.mu

        if lmbda:
            self.lmbda = lmbda
        else:
            self.lmbda = 1 / np.sqrt(np.max(self.D.shape))

    @staticmethod
    def frobenius_norm(M):
        return np.linalg.norm(M, ord='fro')

    @staticmethod
    def shrink(M, tau):
        return np.sign(M) * np.maximum((np.abs(M) - tau), np.zeros(M.shape))

    def svd_threshold(self, M, tau):
        U, S, V = svd(M, full_matrices=False)
        return np.dot(U, np.dot(np.diag(self.shrink(S, tau)), V))

    def fit(self, tol=None, max_iter=1000, iter_print=100):
        iter = 0
        err = np.Inf
        Sk = self.S
        Yk = self.Y
        Lk = np.zeros(self.D.shape)

        if tol:
            _tol = tol
        else:
            _tol = 1E-7 * self.frobenius_norm(self.D)

        while (err > _tol) and iter < max_iter:
            Lk = self.svd_threshold(
                self.D - Sk + self.mu_inv * Yk, self.mu_inv)
            Sk = self.shrink(
                self.D - Lk + (self.mu_inv * Yk), self.mu_inv * self.lmbda)
            Yk = Yk + self.mu * (self.D - Lk - Sk)
            err = self.frobenius_norm(self.D - Lk - Sk)
            iter += 1
            if (iter % iter_print) == 0 or iter == 1 or iter > max_iter or err <= _tol:
                print('iteration: {0}, error: {1}'.format(iter, err))

        self.L = Lk
        self.S = Sk
        return Lk, Sk

    def plot_fit(self, size=None, tol=0.1, axis_on=False):

        n, d = self.D.shape

        if size:
            nrows, ncols = size
        else:
            sq = np.ceil(np.sqrt(n))
            nrows = int(sq)
            ncols = int(sq)

        ymin = np.nanmin(self.D)
        ymax = np.nanmax(self.D)
        print('ymin: {0}, ymax: {1}'.format(ymin, ymax))

        numplots = np.min([n, nrows * ncols])
        plt.figure()

        for n in range(numplots):
            plt.subplot(nrows, ncols, n + 1)
            plt.ylim((ymin - tol, ymax + tol))
            plt.plot(self.L[n, :] + self.S[n, :], 'r')
            plt.plot(self.L[n, :], 'b')
            if not axis_on:
                plt.axis('off')


```

```python
size = 184
img = Image.open("joconde.jpeg")
img = np.array(img)

gray = gray =  0.2989 * np.array(img)[:, :, 0] + 0.5870 * np.array(img)[:, :, 1] + 0.1140 * np.array(img)[: , :, 2]
gray = gray[:size, :size]
rank = 5
```

```python
U, sr, V = truncated_svd(gray, rank)
S_r = np.diag(sr)
image_r = np.dot(U, np.dot(S_r, V))
```

```python
proportion = 0.9
u, v = image_r.shape
mat_mask = mask(u, v, proportion)
```

```python
plt.imshow(image_r)
```

```python
image_raw = mat_mask * image_r
plt.imshow(image_raw)
```

```python
R = R_pca(image_raw)
R.fit()
R.plot_fit()
```

```python
plt.figure(figsize = (15, 10))
plt.subplot(1,3,1)
plt.imshow(image_raw, cmap=plt.cm.gray)
plt.title("original image (rank = {})".format(rank))
plt.subplot(1,3,2)
plt.imshow(R.L, cmap=plt.cm.gray)
plt.title("LOW RANKING IMAGE (proportion = {} per cent)".format(proportion))
plt.subplot(1,3,3)
plt.imshow(R.S, cmap=plt.cm.gray)
plt.title("SPARSE IMAGE")
```
